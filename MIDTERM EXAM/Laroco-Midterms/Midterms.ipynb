```python
#improved version
def translate(text, mapping):
     # Check if inputs are correct types
    if type(text) != str:
        print("Error: text must be a string.")
        return "Invalid input"

    if type(mapping) != dict:
        print("Error: mapping must be a dictionary.")
        return "Invalid input"
        
    result_list = []  # use list for efficient concatenation
    for ch in text.lower():
        if ch in mapping:
            result_list.append(mapping[ch])  # add to list instead of string
        else:
            result_list.append(ch)

    # Join all characters into one string at once
    return "".join(result_list)

    # Strings in Python are immutable, meaning they cannot be changed directly.
    # Every time we use += with a string, Python creates a new string in memory.
    # To make it faster, i use a list (mutable) to collect characters,
    # then join them all at once using "".join(), this is more efficient :>.

# Test case
mapping = {'a':'4','e':'3','i':'1','o':'0','s':'5','t':'7'}
print(translate("I like to eat eggs and spam.", mapping))
```

```python
#improved version
def check_password(pw):

    # Rule checks (all start as False)
    has_uppercase = False
    has_lowercase = False
    has_digit = False
    has_symbol = False
    has_no_spaces = True

    # Check each character in the password
    for ch in pw:
        if ch.isupper():
            has_uppercase = True
        elif ch.islower():
            has_lowercase = True
        elif ch.isdigit():
            has_digit = True
        elif ch in "!@#$%^&*":
            has_symbol = True
        elif ch == " ":
            has_no_spaces = False

    # Check overall length
    long_enough = len(pw) >= 10

    # Determine if password is valid
    if has_uppercase and has_lowercase and has_digit and has_symbol and has_no_spaces and long_enough:
        is_valid = True
    else:
        is_valid = False

    # Return results as a dictionary (Lesson 5–6 topic)
    rules = {
        "Length (>=10)": long_enough,
        "Has Uppercase": has_uppercase,
        "Has Lowercase": has_lowercase,
        "Has Digit": has_digit,
        "Has Symbol": has_symbol,
        "No Spaces": has_no_spaces,
        "Valid Password": is_valid
    }

    return rules


# --- Main Program Loop ---
while True:
    pw = input("Enter your password (type Q to quit): ")

    # Allow user to exit
    if pw.lower() == "q":
        print("You chose to quit. Program ended.")
        break

    result = check_password(pw)

    # Display results
    print("\nPassword Check Results:")
    for rule, status in result.items():
        print(rule + ":", status)

    if result["Valid Password"]:
        print("Password accepted! Good job!")
        break
    else:
        print("❌ Password not accepted. Try again.\n")
```

```python
#improved version
class Vehicle:

    def __init__(self, color, mileage, fuel_liters):
        # Initialize the vehicle's attributes
        self.color = color
        self.mileage = mileage
        self.fuel_liters = fuel_liters

    def drive(self, km, km_per_liter):
        # Check for invalid values (no need to use raise)
        if km < 0 or km_per_liter <= 0:
            print("Invalid distance or efficiency.")
            return 0

        # Calculate fuel needed and distance possible
        fuel_needed = km / km_per_liter

        if self.fuel_liters >= fuel_needed:
            self.mileage += km
            self.fuel_liters -= fuel_needed
            actual_driven = km
        else:
            # Not enough fuel, drive as far as possible
            actual_driven = self.fuel_liters * km_per_liter
            self.mileage += actual_driven
            self.fuel_liters = 0

        return actual_driven

    def refuel(self, liters):
        """Add fuel to the tank."""
        if liters < 0:
            print("Cannot refuel a negative amount.")
        else:
            self.fuel_liters += liters


# Car class inherits from Vehicle
class Car(Vehicle):
    """Car uses the same drive behavior as Vehicle."""
    pass


# Truck class inherits from Vehicle
class Truck(Vehicle):
    """Truck is less fuel-efficient, so it drives shorter distances."""

    def drive(self, km, km_per_liter):
        # Truck consumes 20% more fuel (less efficient)
        adjusted_efficiency = km_per_liter * 0.8
        return Vehicle.drive(self, km, adjusted_efficiency)


# --- Test Section ---
car = Car("Red", 1000, 20)
driven = car.drive(100, 10)
print("Car Mileage:", car.mileage)
print("Car Fuel Left:", car.fuel_liters)
print("Distance Driven:", driven)

truck = Truck("Blue", 500, 50)
truck.drive(200, 10)
print("Truck Mileage:", truck.mileage)
print("Truck Fuel Left:", truck.fuel_liters)
```

### Critique:

Robustness: remove_dog should handle missing names gracefully; oldest_dog should guard empty kennel.

Readability: Use comments and method names per lecture style.

```python
#improved version
# Dog class represents a single dog
class Dog:
    def __init__(self, name, age, breed):
        # Initialize dog attributes
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self):
        # Simple method that returns a string
        return self.name + " says Woof!"


# Kennel class manages multiple Dog objects
class Kennel:

    def __init__(self):
        # Start with an empty list of dogs
        self.dogs = []

    def add_dog(self, dog):
        self.dogs.append(dog)

    def remove_dog(self, name):
        found = False
        new_list = []

        for d in self.dogs:
            if d.name != name:
                new_list.append(d)
            else:
                found = True

        self.dogs = new_list
        return found  # True if a dog was removed

    def oldest_dog(self):
        if len(self.dogs) == 0:
            return None

        oldest = self.dogs[0]
        for d in self.dogs:
            if d.age > oldest.age:
                oldest = d

        return oldest

    def find_by_breed(self, breed):
        result = []
        for d in self.dogs:
            if d.breed.lower() == breed.lower():
                result.append(d)
        return result


# --- Test Program ---
kennel = Kennel()

# Add dogs
dog1 = Dog("Buddy", 5, "Labrador")
dog2 = Dog("Charlie", 8, "Poodle")
dog3 = Dog("Max", 3, "Labrador")

kennel.add_dog(dog1)
kennel.add_dog(dog2)
kennel.add_dog(dog3)

# Display all dogs
print("All Dogs in Kennel:")
for d in kennel.dogs:
    print(d.name, "-", d.age, "years old -", d.breed)

# Find the oldest dog
oldest = kennel.oldest_dog()
print("\nOldest Dog:", oldest.name, "-", oldest.age, "years old")

# Find by breed
labradors = kennel.find_by_breed("Labrador")
print("\nLabrador Dogs:")
for d in labradors:
    print(d.name)

# Remove a dog
removed = kennel.remove_dog("Charlie")
if removed:
    print("\nCharlie was removed from the kennel.")
else:
    print("\nDog not found.")
```
